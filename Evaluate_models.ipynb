### process for Bleu and Bert
need to save the generated texts into files that are formatted into a CSV with a column of input and output

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk import word_tokenize

# Example data
input_texts = ["inp sentence 1", "inp sentence 2"]
output_texts = ["output sentence 1", "output sentence 2"]

# Compute BLEU-1 and BLEU-2
bleu1_scores = []
bleu2_scores = []

for inp, out in zip(input_texts, output_texts):
    reference = [word_tokenize(inp)]
    candidate = word_tokenize(out)
    bleu1 = sentence_bleu(reference, candidate, weights=(1, 0, 0, 0), smoothing_function=SmoothingFunction().method1)
    bleu2 = sentence_bleu(reference, candidate, weights=(0.5, 0.5, 0, 0), smoothing_function=SmoothingFunction().method1)
    bleu1_scores.append(bleu1)
    bleu2_scores.append(bleu2)

print(f"Average BLEU-1: {sum(bleu1_scores) / len(bleu1_scores)}")
print(f"Average BLEU-2: {sum(bleu2_scores) / len(bleu2_scores)}")
```

```python
from bert_score import score

# Assuming output_texts are the generated texts and input_texts are the references
P, R, F1 = score(output_texts, input_texts, lang="en", rescale_with_baseline=True)

print(f"Average Precision: {P.mean().item()}")
print(f"Average Recall: {R.mean().item()}")
print(f"Average F1: {F1.mean().item()}")
```
